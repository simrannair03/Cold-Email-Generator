````python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://jobs.nike.com/job/R-33460")
page_data = loader.load().pop().page_content
print(page_data)
````

````python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
        """
        ### SCRAPED TEXT FROM WEBSITE:
        {page_data}
        ### INSTRUCTION:
        The scraped text is from the career's page of a website.
        Your job is to extract the job postings and return them in JSON format containing the
        following keys: `role`, `experience`, `skills` and `description`.
        Only return the valid JSON.
        ### VALID JSON (NO PREAMBLE):
        """
)

chain_extract = prompt_extract | llm
res = chain_extract.invoke(input={'page_data':page_data})
type(res.content)
````

````python
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
import json

# 1. Create a more structured prompt
prompt = ChatPromptTemplate.from_messages([
    ("system", """You are a job information extractor that outputs STRICT JSON.
    For any missing fields, provide reasonable values rather than null."""),
    ("human", """Extract exactly these fields from the job listing:
{{
    "role": "<job_title>",
    "experience": "<years_or_requirements>",
    "skills": ["<skill1>", "<skill2>"],
    "description": "<job_summary>"
}}

Input text: {text}

Output ONLY the JSON object, nothing else.""")
])

# 2. Create a processing chain
chain = prompt | llm

# 3. Invoke with error handling
try:
    output = chain.invoke({"text": res.content})

    # 4. Manually verify and parse
    print("Raw LLM output:\n", output.content)  # Debugging

    # Try to extract JSON from the output (handling markdown or other wrappers)
    json_str = output.content.strip()
    if json_str.startswith("```json"):
        json_str = json_str[7:-3]  # Remove markdown wrapper
    elif json_str.startswith("```"):
        json_str = json_str[3:-3]

    parsed = json.loads(json_str)
    print("Parsed result:", parsed)

except json.JSONDecodeError as e:
    print(f"Failed to parse JSON: {e}")
    print("Raw output was:", output.content)
except Exception as e:
    print(f"Error: {e}")
````
